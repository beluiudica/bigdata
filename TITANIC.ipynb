```python
import sqlite3 # a module in Python's standard library that provides an interface to the SQLite database.
import pandas as pd
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_data = pd.read_csv(url)
conn = sqlite3.connect(':memory:')
titanic_data.to_sql('titanic', conn, index=False)
# Function to run querys
def run_query(query, connection=conn):
    return pd.read_sql_query(query, connection)
```

```python
#Veo mis datos
titanic_data
```

```python
#Ver los nombres de 10 pasajeros.

# Usando SQL

run_query(
"""
select name
from titanic
limit 10

"""

 )
```

```python
#cuantas clases diferentes hay?
# Usando SQL

run_query(
"""
SELECT COUNT(DISTINCT Pclass) FROM titanic;

"""

 )
```

```python
#cuantas clases diferentes hay?
# Usando Pandas

titanic_data['Pclass'].nunique()
```

```python
#Mostrar los pasajeros ordenados por edad
# Usando SQL
run_query(
"""
SELECT *
FROM titanic
ORDER BY Age;

"""

 )
```

```python
#Mostrar los pasajeros ordenados por edad
# Usando Pandas
titanic_data.sort_values(by='Age')
```

```python
#Mostrar los pasajeros que sobrevivieron
#Usando SQL
run_query(
"""
SELECT *
FROM titanic
WHERE Survived = 1;

"""

 )
```

```python
#Mostrar los pasajeros que sobrevivieron
#Usando Pandas
titanic_data[titanic_data['Survived'] == 1]
```

```python
#Mostrar los pasajeros que no sobrevivieron y tenÃ­an menos de 30 aÃ±os
#Usando SQL
run_query(
"""
SELECT *
FROM titanic
WHERE Survived = 0 AND Age < 30;

"""

 )
```

```python
#Mostrar los pasajeros que no sobrevivieron y tenÃ­an menos de 30 aÃ±os
#Usando Pandas
titanic_data[(titanic_data['Survived'] == 0) & (titanic_data['Age'] < 30)]
```

```python
#Mostrar pasajeron que tienen entre 30 y 50 aÃ±os
#Usando SQL
run_query(
"""
SELECT *
FROM titanic
WHERE Age BETWEEN 30 AND 50;

"""

 )
```

```python
#Mostrar pasajeron que tienen entre 30 y 50 aÃ±os
#Usando Pandas
titanic_data[(titanic_data['Age'] >= 30) & (titanic_data['Age'] <= 50)]
```

```python
#Â¿CuÃ¡ntos pasajeros se llaman 'John'?
#Usando SQL
run_query(
"""
SELECT COUNT(*)
FROM titanic
WHERE Name LIKE '%John%';

"""

 )
```
