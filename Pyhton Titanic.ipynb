```python
#Volvemos a cargar el dataset:
data = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
# Ajustamos la columna "Fare" a tipo entero
data['Fare'] = data['Fare'].astype(int)
```

```python
###2: Elige una columna numérica y crea un boxplot para ella. ¿Ves algún dato atípico?
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot for the 'Fare' column
sns.boxplot(x=data['Fare'])
plt.title('Boxplot of Fare')
plt.show()
```

```python
###3: ¿Cómo filtrarías los datos para mostrar solo a los pasajeros de primera clase que sobrevivieron?
```

```python
# Filter the data for passengers in first class who survived
first_class_survived = data[(data['Pclass'] == 1) & (data['Survived'] == 1)]

# Display the filtered data
print(first_class_survived)
```

```python
###4: ¿Cómo agruparías a los pasajeros por sexo y clase, y calcularías la tasa de supervivencia promedio para cada grupo?
```

```python
# Group by Sex and Pclass and calculate the mean of Survived
survival_rate_by_sex_class = data.groupby(['Sex', 'Pclass'])['Survived'].mean()

# Display the result
print(survival_rate_by_sex_class)
```

```python
###5: ¿Cuál es la edad promedio de los pasajeros que sobrevivieron y de los que no?
```

```python
# Group by Survived and calculate the mean of Age
average_age_by_survival = data.groupby('Survived')['Age'].mean()

# Display the result
print(average_age_by_survival)
```

```python
###6: ¿Existe una correlación significativa entre la edad y la tarifa del pasaje?
```

```python
# Visualize the distribution of Age for survivors and non-survivors using boxplots
sns.boxplot(x='Survived', y='Age', data=data)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

```python
###7: ¿Cuál es la moda de la clase social de los pasajeros?
```

```python
# Calculate the mode of the 'Pclass' column
pclass_mode = data['Pclass'].mode()

# Display the result
print("La moda de la clase social es:", pclass_mode)
```

```python
###9: Contesta con tus palabras: ¿Qué significa que una varibles es categórica o númerica? Menciona algunos ejemplos de cada una
Variables Numéricas: aquellas que representan cantidades y pueden ser medidas en una escala. Sus valores son números y se pueden realizar operaciones matemáticas con ellos.
Ejemplos en el dataset del Titanic: Age (Edad), Fare (Tarifa), Siblings/Spouses Aboard (Hermanos/Cónyuges a bordo), Parents/Children Aboard (Padres/Hijos a bordo).
Variables Categóricas: aquellas que representan cualidades, características o grupos. Sus valores son categorías o etiquetas y no se pueden realizar operaciones matemáticas significativas con ellos (aunque a veces se les asignen números para representarlos, estos números no tienen un valor cuantitativo real).
Ejemplos en el dataset del Titanic: Survived (Sobrevivió - aunque sea 0 o 1, representa una categoría), Pclass (Clase social - 1, 2, 3 son categorías), Sex (Sexo), Name (Nombre - aunque cada nombre es único, es una etiqueta).
En resumen, las variables numéricas son "cuánto" o "cuántos", mientras que las categóricas son "de qué tipo" o "a qué grupo pertenece".
```
